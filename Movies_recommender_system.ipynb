```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
rating_mat = ratings.pivot(index="movieId", columns="userId", values="rating").fillna(0)
```

```python
rating_mat
```

```python
distance_mtx = squareform(pdist(rating_mat, 'cosine'))
similarity_mtx = 1- distance_mtx
```

```python
distance_mtx.shape
```

```python
user_to_predict = 2
item_to_predict = 3

user2_ratings = rating_mat.iloc[:,user_to_predict-1]
item3_similarity = similarity_mtx[item_to_predict-1]

numerator = np.dot(user2_ratings,item3_similarity)
denom = item3_similarity[user2_ratings > 0].sum()
prediction = numerator / denom
print(prediction)
```

```python
def calculate_user_rating(userid, similarity_mtx, rating_mat):
    user_rating = rating_mat.iloc[:,userid-1]
    pred_rating = copy.deepcopy(user_rating)
    
    default_rating = user_rating[user_rating>0].mean()
    numerate = np.dot(similarity_mtx, user_rating)
    corr_sim = similarity_mtx[:, user_rating >0]
    for i,ix in enumerate(pred_rating):
        temp = 0
        if ix < 1:
            w_r = numerate[i]
            sum_w = corr_sim[i,:].sum()
            if w_r == 0 or sum_w == 0:
                temp = default_rating
            else:
                temp = w_r / sum_w
            pred_rating.iloc[i] = temp
    return pred_rating


def recommendation_to_user(userid, top_n, similarity_mtx, rating_mat):
    user_rating = rating_mat.iloc[:,userid-1]
    pred_rating = calculate_user_rating(userid, similarity_mtx, rating_mat)

    top_item = sorted(range(1,len(pred_rating)), key = lambda i: -1*pred_rating.iloc[i])
    top_item = list(filter(lambda x: user_rating.iloc[x]==0, top_item))[:top_n]
    res = []
    for i in top_item:
        res.append(tuple([new[new['movie_id']==ratings['movieId'][i]]['title'].tolist()[0], pred_rating.iloc[i]]))
    
    return res,pred_rating

res,pred_rating = recommendation_to_user(2,10,similarity_mtx, rating_mat)
```
